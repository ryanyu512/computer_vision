```python
from stereo_visual_odometry_r import *
from matplotlib import pyplot as plt
```

# Demo 1 (Straight Line)

```python
vo = VisualOdometry('KITTI_sequence_1')

gt_path = []
est_path = []

for i, gt_pose in enumerate(vo.gt_poses):
    if i < 1:
        c_pose = gt_pose
    else:
        T_mat = vo.get_relative_pose(i)
        c_pose = np.matmul(c_pose, T_mat)
        
    est_path.append((c_pose[0, 3], c_pose[2, 3]))
    gt_path.append((gt_pose[0, 3], gt_pose[2, 3]))


gt_path = np.array(gt_path)
est_path = np.array(est_path)

plt.plot(gt_path[:, 0], gt_path[:, 1], '.r')
plt.plot(est_path[:, 0], est_path[:, 1], 'xb')

plt.xlim([-50, 50])
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['ground truth', 'estimation'])
plt.show()
```
