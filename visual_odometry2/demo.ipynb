```python
from mono_visual_odometry import *
from matplotlib import pyplot as plt
```

# Demo 1 (Straight Line)

```python
vo = VisualOdometry('KITTI_sequence_1')    

gt_path = []
es_path = []

for i, gt_pose in enumerate(vo.gt_poses):
    if i == 0:
        c_pose = gt_pose
    else:
        #q1: key points in previous frame
        #q2: key points in current frame
        q1, q2 = vo.get_matches(i)
        #compute transformation matrix of previous frame relative to current frame
        T_mat  = vo.compute_pose(q1, q2)
        #compute transformation matrix of current frame relative to the world coordinate    
        c_pose = np.matmul(c_pose, np.linalg.inv(T_mat))

    gt_path.append([gt_pose[0, 3], gt_pose[2, 3]])
    es_path.append([c_pose[0, 3], c_pose[2, 3]])
    
gt_path = np.array(gt_path)
es_path = np.array(es_path)

plt.plot(gt_path[:, 0], gt_path[:, 1], '.r')
plt.plot(es_path[:, 0], es_path[:, 1], '.b')

plt.xlim([-50, 50])
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['ground truth', 'estimation'])
plt.show()
```
